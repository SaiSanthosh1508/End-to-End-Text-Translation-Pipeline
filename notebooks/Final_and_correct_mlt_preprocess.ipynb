```python
language_map = {
    "Arabic" : 0,
    "Latin" : 1,
    "Chinese" : 2,
    "Korean" : 3,
    "Japanese" : 4,
    "Bangla" : 5,
    "Hindi" : 6,
    "Other" : 7
}
```

```python
import cv2
from pathlib import Path
import os
import shutil

def convert_to_yolo(images_dir, annots_dir, target_dir):

  Path(target_dir).mkdir(parents=True, exist_ok=True)
  Path(os.path.join(target_dir,"images")).mkdir(parents=True, exist_ok=True)
  Path(os.path.join(target_dir,"labels")).mkdir(parents=True, exist_ok=True)
  Path(os.path.join(target_dir,"images","train")).mkdir(parents=True, exist_ok=True)
  Path(os.path.join(target_dir,"images","val")).mkdir(parents=True, exist_ok=True)
  Path(os.path.join(target_dir,"labels","train")).mkdir(parents=True, exist_ok=True)
  Path(os.path.join(target_dir,"labels","val")).mkdir(parents=True, exist_ok=True)

  images = sorted(os.listdir(images_dir))
  labels = sorted(os.listdir(annots_dir))

  if len(images) != len(labels):
    return "The number of images do not match the label count"

  for i in range(0,len(images),1000):
    image_chunk = images[i:i+1000]
    label_chunk = labels[i:i+1000]

    ## train split
    for j in range(1000):
      image_name = image_chunk[j]
      label_name = label_chunk[j]
      image_path = os.path.join(images_dir,image_name)
      label_path = os.path.join(annots_dir,label_name)

      with open(label_path) as f:
        lines = f.readlines()
        if len(lines) == 0:
          continue

      # image width and image height
      img = cv2.imread(image_path)
      yolo_lines = []

      for line in lines:
        x1,y1,x2,y2,x3,y3,x4,y4,lang,_ = line.split(",",9)

        if lang in language_map:
          lang_id = language_map[lang]
        else:
          lang_id = language_map["Other"]

        x1,y1,x2,y2,x3,y3,x4,y4 = int(x1),int(y1),int(x2),int(y2),int(x3),int(y3),int(x4),int(y4)

        min_x = min(x1,x2,x3,x4)
        max_x = max(x1,x2,x3,x4)
        min_y = min(y1,y2,y3,y4)
        max_y = max(y1,y2,y3,y4)

        bbox_width = max_x - min_x
        bbox_height = max_y - min_y

        center_x = min_x + bbox_width/2
        center_y = min_y + bbox_height/2

        if img is not None:
          img_h,img_w,_ = img.shape
          x_center = center_x/img_w
          y_center = center_y/img_h
          bbox_width = bbox_width/img_w
          bbox_height = bbox_height/img_h

          bbox_line = f"{lang_id} {x_center} {y_center} {bbox_width} {bbox_height}\n"
          yolo_lines.append(bbox_line)

        else:
          print(f"{image_path} image read as None")

      if j < 900:
        with open(os.path.join(target_dir,"labels","train",label_name),"w") as f:
          f.writelines(yolo_lines)

        shutil.copyfile(image_path,os.path.join(target_dir,"images","train",image_name))
      else:
        with open(os.path.join(target_dir,"labels","val",label_name),"w") as f:
          f.writelines(yolo_lines)

        shutil.copyfile(image_path,os.path.join(target_dir,"images","val",image_name))

  print("Successfully created YOLO dataset")
```

```python
convert_to_yolo(
    images_dir = "/content/images",
    annots_dir = "/content/train_gt",
    target_dir = "/content/mlt_yolo_final"
)
```
